### Explore the missing values

```python
# what is the number of missing values present 
train.columns[train.isnull().sum()>0] # columns with missing value present 
```

```python
# Having identified columns with missing values above, we decide to do away with some which have a certain percent of missing values 
# let us get this percentages 

total_missing=train.isnull().sum().sort_values(ascending=False)
        #isnull() checks each cell in the DF and returns True if it's a missing value (NaN), and False if it's not.
        #sum() calculates the sum of True values for each column
        #sort_values sorts the counts of missing values in descending order, so the columns with the most missing values are at the top

percent=train.isnull().sum().sort_values(ascending=False)/train.isnull().count().sort_values(ascending=False)
        # the above code contains the % of missing values for each column, sorted in descending order
        #count() calculates the total number of non-missing values for each column.
        # The division calculates the ratio of missing values to total values giving us %

missing_df=pd.concat([total_missing,percent],axis=1,keys=['total_missing','percentage_missing'],sort=False)
        # Above code creates a DF by concatenating the two Series along the columns axis.
        # axis=1 means that the concat is done horizontally (along columns)
        # keys assigns column names to the concatenated Series.
        # sort=False ensures that columns are not sorted.

missing_df.index.name='variable'# set the index name to variable 

missing_df=missing_df[total_missing>0] # filters dataset to only retain those with missing values 

missing_df['percentage_missing']=missing_df['percentage_missing'].map(lambda x:x*100) 
        # convert to percentage 

missing_df
```

- It is above the Third Quadrant (Q3) added by (1.5 x IQR)

```python
from sklearn import tree
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeRegressor
from sklearn.linear_model import LinearRegression
```

```python
# Selecting columns I am interested in
columns = ['grlivarea' , 'garagearea' , 'totalbsmtsf' , '1stflrsf', 'saleprice']
df = train.loc[:, columns]

df.head()
```

```python
# Splitting the data
# features I could use = ['grlivarea' , 'garagearea' , 'totalbsmtsf' , '1stflrsf']
X = train.loc[:, ['grlivarea']]
y = train.loc[:, ['saleprice']]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0, train_size = .75)
```

```python
#fitting simple linear regression to the training set
lr = LinearRegression ()
lr.fit (X_train, y_train) #--> We call the fit method if we want to train the dataset
```

```python
#Predicting the Test set results
y_predict = lr.predict(X_test) #Predict takes the independent variable as it is predicting the dependable variable
y_predict #this output is what the model has predicted
        #We shall compare the y_prediction to the y_test to see the accuracy of what has been predicted
```

```python
#Visualizing the training set results
plt.scatter (X_train, y_train, color='brown')
plt.plot (X_train, lr.predict(X_train), color = 'green')

plt.title ('Housing sale price prediction based on ground flr area')
plt.xlabel ('Ground floor living area')
plt.ylabel ('Sale Price')
plt.show ()
```
